```python
from selenium import webdriver
import os
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import time
from selenium.webdriver.common.keys import Keys
import pandas as pd 
import random 
from parsel import Selector
import openpyxl
from openpyxl.worksheet.table import Table
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver import ActionChains
import requests
```

```python
class MDIIMG(webdriver.Chrome):
    def __init__(self, profile_path=None, driver_path =r"C:\Users\Mohsin\Documents\Selenium"):
        self.profile_path = profile_path
        self.driver_path = driver_path
        os.environ['PATH'] += self.driver_path
        super(MDIIMG, self).__init__(options=self._get_chrome_options())
        
    def _get_chrome_options(self):
        chrome_options = webdriver.ChromeOptions()
        if self.profile_path:
            chrome_options.add_argument(f"user-data-dir={self.profile_path}")
            
        user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
        chrome_options.add_argument(f"user-agent={user_agent}")
        return chrome_options    
        
    def home_page(self):
        wait = WebDriverWait(self,10)
        self.get('https://www.appfolio.com/')
        
    def login_button(self):
        login_element = self.find_element(By.CLASS_NAME, "login")
        actions = ActionChains(self)
        actions.move_to_element(login_element).perform()
        
    def property_manager_login(self):
        wait = WebDriverWait(self,10)
        property_manager_login_element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'a[href="https://passport.appf.io/sign_in?idp_type=property&vhostless=true"]')))
        property_manager_login_element.click()
        
    def login_details(self, username, password):
        WebDriverWait(self, 10).until(EC.number_of_windows_to_be(2))  
        self.switch_to.window(self.window_handles[-1])
        username_element = self.find_element(By.ID, "usernameInput")
        username_element.send_keys(username)
        password_element = self.find_element(By.ID, "passwordInput")
        password_element.send_keys(password)
        submit_button = self.find_element(By.CSS_SELECTOR, '[type="submit"]')
        submit_button.click()
         
    def click_report(self):
        wait = WebDriverWait(self,10)
        click_report_element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, '[href="#Reporting"]')))
        click_report_element.click()
        
    def click_reports(self):
        click_reports_element = self.find_element(By.CSS_SELECTOR, '[href="/reports"]')
        click_reports_element.click()
        
    def delinquency(self):
        wait = WebDriverWait(self,10)
        delinquency_element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'a[href="https://mdesignsinvestments.appfolio.com/buffered_reports/delinquency_as_of?customize=true"]')))
        delinquency_element.click()
        
    def delinquency_options(self):
        WebDriverWait(self, 10).until(EC.number_of_windows_to_be(3))  
        self.switch_to.window(self.window_handles[-1])
        wait = WebDriverWait(self, 10)
        delinquency_options_element = wait.until(EC.element_to_be_clickable((By.XPATH, '//span[@class="Select-clear-zone"]')))
        delinquency_options_element.click()
        
    def choose_property(self):
        choose_property_element = self.find_element(By.CSS_SELECTOR, 'a[class="js-button-columns nav-link"]')
        choose_property_element.click()
        tags_checkbox = self.find_element(By.XPATH, '//span[text()="Tags"]/preceding-sibling::input[@type="checkbox"]')
        tags_checkbox.click()
        tenant_status = self.find_element(By.XPATH, '//span[text()="Tenant Status"]/preceding-sibling::input[@type="checkbox"]')
        tenant_status.click()
        property_checkbox = self.find_element(By.XPATH, '//span[text()="Property"]/preceding-sibling::input[@type="checkbox"]')
        property_checkbox.click()
        update_button = self.find_element(By.CSS_SELECTOR, 'button[class="mr-2 btn btn-primary"]')
        update_button.click()
        
    def action_export(self):
        time.sleep(10)
        action_export_element = self.find_element(By.XPATH, '//span[text()="Actions"]')
        self.execute_script("arguments[0].click();", action_export_element)
        
    def export_excel(self):
        export_excel_element = self.find_element(By.CSS_SELECTOR, 'button[class="js-export-excel-button dropdown-item"]')
        export_excel_element.click()
    
    def previous_page(self):
        WebDriverWait(self, 10).until(EC.number_of_windows_to_be(3))  
        self.switch_to.window(self.window_handles[-2])
        time.sleep(5)
        unit_vacancy = self.find_element(By.CSS_SELECTOR, 'a[href="https://mdesignsinvestments.appfolio.com/buffered_reports/unit_vacancy?customize=true"]')
        unit_vacancy.click()
   
    def updatee_button(self):
        WebDriverWait(self, 10).until(EC.number_of_windows_to_be(4))  
        self.switch_to.window(self.window_handles[-1])
        #wait = WebDriverWait(self,10)
        updatee_button_element = self.find_element(By.CSS_SELECTOR, 'button[class="mr-2 btn btn-primary"]')
        updatee_button_element.click()
        
    def actions_export(self):
        time.sleep(10)
        action_export_element = self.find_element(By.XPATH, '//span[text()="Actions"]')
        self.execute_script("arguments[0].click();", action_export_element)
        export_excel_element = self.find_element(By.CSS_SELECTOR, 'button[class="js-export-excel-button dropdown-item"]')
        export_excel_element.click()
        
    def previouss_page(self):
        WebDriverWait(self, 10).until(EC.number_of_windows_to_be(4))
        self.switch_to.window(self.window_handles[-3])
        time.sleep(5)
        work_orders = self.find_element(By.CSS_SELECTOR, 'a[href="https://mdesignsinvestments.appfolio.com/memorized_reports/139"]')
        work_orders.click()
        
    def actionss_export(self):
        WebDriverWait(self, 10).until(EC.number_of_windows_to_be(5))  
        self.switch_to.window(self.window_handles[-1])
        time.sleep(5)
        actions_export_element = self.find_element(By.XPATH, '//span[text()="Actions"]')
        self.execute_script("arguments[0].click();", actions_export_element)
        exports_excel_element = self.find_element(By.CSS_SELECTOR, 'button[class="js-export-excel-button dropdown-item"]')
        exports_excel_element.click()
        
    def next_next_page(self):
        next_element = self.find_element(By.XPATH, '')
        
        
```
